```python
import itertools
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
def convert(x):
    ids = np.unique(x)
    vals = 0
    pd = np.zeros((x.shape))
    for item in ids:
        pd[x==item] = vals
    return pd
```

```python
train = pd.read_csv("train.csv")
train.head()
```

```python
test = pd.read_csv("test.csv")
print(test.shape)
test.head()
```

```python
train1 = pd.read_csv("train1.csv")
train1.head()
```

```python
train2 = pd.read_csv("train2.csv")
train2.head()
```

```python
train3 = pd.read_csv("train3.csv")
train3.head()
```

```python
X1 = train1.drop(['Y'],axis=1)
X1['ID'] = convert(X1['ID'].values)

test['ID'] = convert(test['ID'].values)

X1 = X1.apply(pd.to_numeric, errors='coerce')
X1.fillna(0, inplace = True)

test = test.apply(pd.to_numeric, errors='coerce')
test.fillna(0, inplace = True)
X1 = X1.astype(float)
test = test.astype(float)
pd.options.display.max_columns = None
```

```python
X2 = train2.drop(['Y'],axis=1)
X2['ID'] = convert(X2['ID'].values)
test['ID'] = convert(test['ID'].values)
X2 = X2.apply(pd.to_numeric, errors='coerce')
X2.fillna(0, inplace = True)
test = test.apply(pd.to_numeric, errors='coerce')
test.fillna(0, inplace = True)
X2 = X2.astype(float)
test = test.astype(float)
pd.options.display.max_columns = None
```

```python
X3 = train3.drop(['Y'],axis=1)
X3['ID'] = convert(X3['ID'].values)
test['ID'] = convert(test['ID'].values)
X3 = X3.apply(pd.to_numeric, errors='coerce')
X3.fillna(0, inplace = True)
test = test.apply(pd.to_numeric, errors='coerce')
test.fillna(0, inplace = True)
X3 = X3.astype(float)
test = test.astype(float)
pd.options.display.max_columns = None
```

```python
X = train.drop(['Y'],axis=1)
X['ID'] = convert(X['ID'].values)

test['ID'] = convert(test['ID'].values)

X = X.apply(pd.to_numeric, errors='coerce')
X.fillna(0, inplace = True)

test = test.apply(pd.to_numeric, errors='coerce')
test.fillna(0, inplace = True)

y = train['Y']

X = X.astype(float)
test = test.astype(float)
pd.options.display.max_columns = None
```

```python
y=train['Y']
y1=train1['Y']
y2=train2['Y']
y3=train3['Y']
y,y1,y2,y3
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20,random_state=
                                                    12)
```

```python
from xgboost import XGBClassifier
from xgboost import plot_importance
from sklearn import metrics
clf = XGBClassifier(n_estimators=700,
        # 如同學習率
        learning_rate= 0.1, 
       )
clf.fit(X_train, y_train)
test_predict=clf.predict(X_test)
```

```python
test_predict1=clf.predict(X1)
test_predict2=clf.predict(X2)
test_predict3=clf.predict(X3)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, test_predict)
score1 = accuracy_score(y1, test_predict1)
score2 = accuracy_score(y2, test_predict2)
score3 = accuracy_score(y3, test_predict3)
score,score1,score2,score3
```
