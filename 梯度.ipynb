```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
#pip install xgboost
```

```python
train = pd.read_csv("train.csv")
print(train.shape)
train.head()
```

```python
test = pd.read_csv("test.csv")
print(test.shape)
test.head()
```

```python
def convert(x):
    ids = np.unique(x)
    vals = 0
    pd = np.zeros((x.shape))
    for item in ids:
        pd[x==item] = vals
    return pd
X = train.drop(['Y'],axis=1)
X['ID'] = convert(X['ID'].values)

test['ID'] = convert(test['ID'].values)

X = X.apply(pd.to_numeric, errors='coerce')
X.fillna(0, inplace = True)

test = test.apply(pd.to_numeric, errors='coerce')
test.fillna(0, inplace = True)

y = train['Y']

X = X.astype(float)
test = test.astype(float)
pd.options.display.max_columns = None
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20,random_state=0)
y_test
```

```python
from sklearn.ensemble import GradientBoostingClassifier 
from sklearn import metrics
from  sklearn.datasets  import  make_hastie_10_2
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from xgboost import plot_importance
from sklearn import metrics
clf = GradientBoostingClassifier(loss="deviance", learning_rate=0.1, n_estimators=150)
#clf=  XGBClassifier(n_estimators=600,learning_rate= 0.1)
clf.fit(X_train, y_train)
y_pred = clf.predict(test)
y_pred
results = np.around(clf.predict_proba(test),2)
results.transpose()
```

```python
res = pd.DataFrame(results, columns = ['C1','C2','C3'])
df = pd.DataFrame(data=res)
df.to_csv('g.csv')
```
