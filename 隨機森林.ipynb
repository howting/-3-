```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
train = pd.read_csv("train.csv")
print(train.shape)
train.head()
```

```python
test = pd.read_csv("test.csv")
print(test.shape)
test.head()
```

```python
def convert(x):
    ids = np.unique(x)
    vals = 0
    pd = np.zeros((x.shape))
    for item in ids:
        pd[x==item] = vals
    return pd
X = train.drop(['Y'],axis=1)
X['ID'] = convert(X['ID'].values)

test['ID'] = convert(test['ID'].values)

X = X.apply(pd.to_numeric, errors='coerce')
X.fillna(0, inplace = True)

test = test.apply(pd.to_numeric, errors='coerce')
test.fillna(0, inplace = True)

y = train['Y']

X = X.astype(float)
test = test.astype(float)
pd.options.display.max_columns = None
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,random_state=0)
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import  preprocessing, metrics
et_clf = RandomForestClassifier(n_estimators = 100)
#et_clf = et_clf.ensemble.RandomForestClassifi0r(n_estimators = 1000)
#Train model on training data-
et_clf.fit(X_train, y_train)

# Make predictions using trained model-
y_pred = et_clf.predict(test)
y_pred
results = np.around(et_clf.predict_proba(test),2)
results.transpose()
res = pd.DataFrame(results, columns = ['C1','C2','C3'])
df = pd.DataFrame(data=res)
df.to_csv('A.csv')
```
